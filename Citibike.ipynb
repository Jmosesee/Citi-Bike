```python
def to_dt(s):
    try:
        return datetime.strptime(s, '%Y-%m-%d %H:%M:%S.%f')
    except:
        try:
            return datetime.strptime(s, '%Y-%m-%d %H:%M:%S')
        except:
            raise
```

```python
def from_dt(dt):
    return datetime.strftime(dt, '%Y-%m-%d %H:%M:%S.%f')
```

```python
c = df['created_at']
f = []
previous_dt = datetime.min
previous_i = 0
for i, v in tqdm(c.iteritems()):
    try:
        dt = to_dt(v)
        if previous_dt > dt:
            df.loc[i,'created_at'] = from_dt(to_dt(df.loc[previous_i,'created_at']) + timedelta(microseconds=1))
            f.append(previous_i)
            f.append(i)
        previous_i = i
        previous_dt = dt
    except:
        print (i)
        raise
```

```python
start_date = datetime.strptime('2014-01', '%Y-%m')
end_date = datetime.strptime('2018-10', '%Y-%m')
date = start_date
dates = [date]
while (date < end_date):
    date = date + relativedelta(months=1)
    dates.append(date)
date_strings = [datetime.strftime(d, '%Y-%m') for d in dates]
date_tuples = zip(date_strings, date_strings[1:])
```

```python
y = 14
m = 1
path_list = []
while y < 18 or m <= 9: 
    m_folder_name = f'20{y:02}{m:02}-citibike-tripdata'
    if (y*12+m) < (14*12+9):
        m_file_name = f'20{y:02}-{m:02} - Citi Bike trip data.csv'
    else:
        m_file_name = m_folder_name + '.csv'
    if y >= 17:
        m_folder_name += '.csv'
    if (y*12+m) >= (15*12+9):
        if y==17 and m==8:
            JC_file_name = f'JC-20{y:02}{m:02} citibike-tripdata.csv'
        else:
            JC_folder_name = f'JC-20{y:02}{m:02}-citibike-tripdata.csv'
        if (y*12+m) < (16*12+4) and (y*12+m) > (15*12+9):
            JC_file_name = f'JC-20{y:02}{m}-citibike-tripdata.csv'
        else:
            JC_file_name = JC_folder_name
        path_list.append((m_folder_name, m_file_name, JC_folder_name, JC_file_name))
    else:
        path_list.append((m_folder_name, m_file_name))
    m += 1
    if m >= 13:
        m = 1
        y += 1
```

```python
import os
for path in path_list:
    if len(path) > 2:
        if (os.path.isfile('C:\\Users\\jmose\\Downloads\\' + path[2] + '\\' + path[3])):
            print ("OK")
        else:
            print ('C:\\Users\\jmose\\Downloads\\' + path[2] + '\\' + path[3])
```

```python
hourly_aggregation_monthly_list = []
original_columns = []
for path in tqdm(path_list):
#     path = path_list[34]
    df = pd.read_csv('C:\\Users\\jmose\\Downloads\\' + path[0] + '\\' + path[1])
    if len(original_columns) == 0:
        original_columns = df.columns
    else:
        df.columns = original_columns
#     if 'Start Time' in df.columns:
#         df = df.rename(columns = {'Start Time': 'starttime', 'Stop Time': 'stoptime', 'User Type': 'usertype', 'Gender': 'gender'})
    df["starttime"] = pd.to_datetime(df['starttime'])
    df["stoptime"] = pd.to_datetime(df['stoptime'])
    if len(path) > 2:
        dfJC = pd.read_csv('C:\\Users\\jmose\\Downloads\\' + path[2] + '\\' + path[3])
        dfJC.columns = original_columns
#         if 'Start Time' in dfJC.columns:
#             dfJC = dfJC.rename(columns = {'Start Time': 'starttime', 'Stop Time': 'stoptime', 'User Type': 'usertype', 'Gender': 'gender'})
        dfJC["starttime"] = pd.to_datetime(dfJC['starttime'])
        dfJC["stoptime"] = pd.to_datetime(dfJC['stoptime'])
        dfc = pd.concat([df.set_index('starttime'), dfJC.set_index('starttime')], sort=True)
    else:
        dfc = df.set_index('starttime').sort_index()
    hourly_aggregation_monthly_list.append(
        dfc.groupby(['usertype', 'gender'] + [pd.Grouper(freq = 'H')]).size().reset_index()
    )
```

```python
df = pd.concat(hourly_aggregation_monthly_list)
```

```python
df.to_csv('hourly_aggregation.csv', index=False)
```
